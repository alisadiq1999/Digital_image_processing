```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('Capture.PNG',0)
print('Original Img: ')
cv2_imshow(img)

structuringElement = np.ones([5,5])

erodedImg = cv2.erode(img, structuringElement, iterations = 1)
finalImg1 = cv2.dilate(erodedImg, structuringElement, iterations = 1)

print('Erosion + Dialation: ')
cv2_imshow(finalImg1)
cv2.imwrite('Task1Img1.jpg', finalImg1)

dilatedImg = cv2.dilate(img, structuringElement, iterations = 1)
finalImg2 = cv2.erode(dilatedImg, structuringElement, iterations = 1)

print('Dialation + Erosion')
cv2_imshow(finalImg2)
cv2.imwrite('Task1Img2.jpg', finalImg2)
```

```
#Task 2

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('Capture2.PNG', 0) 

print("Original Image:")
cv2_imshow(img)
(height, width) = img.shape

ret, thresh = cv2.threshold(img,127,255,cv2.THRESH_BINARY)

dilateSE = np.ones((1,5), np.uint8)  
erodeSE =  np.ones((7,6), np.uint8)

dilatedImg = cv2.dilate(thresh, dilateSE, iterations=1) 
erodedImg = cv2.erode(dilatedImg, erodeSE, iterations = 1) 

print("Morphological Operations:")
cv2_imshow(erodedImg) 
cv2.imwrite('Task2Img1.jpg', erodedImg)

cannyImg = np.zeros([height,width])
cannyImg = cv2.Canny(erodedImg,40,130)

print("Canny Operator:")
cv2_imshow(cannyImg)
cv2.imwrite('Task2Img2.jpg', cannyImg)
```
